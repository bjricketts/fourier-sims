```python
import numpy as np
import matplotlib.pyplot as plt

# Set up figure with equal aspect ratio
fig, ax = plt.subplots(figsize=(7, 7))

# Axis limits (symmetric)
lim = 5
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_aspect('equal')

# Shade the upper half-plane in grey
ax.fill_between([-lim, lim], 0, lim, color='lightgrey', alpha=0.6,
                label='Upper Half-Plane (Im(ω) > 0)')

# Draw axes
ax.axhline(0, color='black', linewidth=1.0)
ax.axvline(0, color='black', linewidth=1.0)

# Three valid points in the upper half-plane (Im > 0)
points = [
    (2.0, 1.5, '$z_1 = 2 + 1.5i$'),
    (-1.5, 3.0, '$z_2 = -1.5 + 3i$'),
    (3.5, 0.8, '$z_3 = 3.5 + 0.8i$'),
]

for re, im, label in points:
    ax.plot(re, im, 'o', color='crimson', markersize=10,
            markeredgecolor='black', markeredgewidth=1.2, zorder=5)
    ax.annotate(label, xy=(re, im), xytext=(re + 0.25, im + 0.25),
                fontsize=11)

# Labels for the half-planes
ax.text(0, lim - 0.4, 'Upper Half-Plane', ha='center',
        fontsize=12, style='italic', color='dimgrey')
ax.text(0, -lim + 0.25, 'Lower Half-Plane', ha='center',
        fontsize=12, style='italic', color='dimgrey')

# Axis labels
ax.set_xlabel('Re(ω)', fontsize=12)
ax.set_ylabel('Im(ω)', fontsize=12)
ax.set_title('The Complex ω-Plane', fontsize=13)

# Grid
ax.grid(True, alpha=0.3, linestyle='--')

# Legend
#ax.legend(loc='lower right', framealpha=0.9)

plt.tight_layout()
plt.savefig("figures/uhp_plot.png", dpi=120, bbox_inches='tight')
plt.show()
print("Plot saved.")
```
